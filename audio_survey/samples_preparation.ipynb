```python
# NORMALIZE LOUDNESS

import soundfile as sf
import pyloudnorm as pyln

filename = "../csv/new_2/03_70.wav"

data, rate = sf.read(filename) # load audio

# measure the loudness first 
meter = pyln.Meter(rate) # create BS.1770 meter
loudness = meter.integrated_loudness(data)

# loudness normalize audio to -23 dB LUFS
loudness_normalized_audio = pyln.normalize.loudness(data, loudness, -23.0)

sf.write(filename, loudness_normalized_audio, rate)
```

```python
# TRIMMING file

from pydub import AudioSegment

refname = "new_2/03.wav"
ref = AudioSegment.from_wav(refname)
l = len(ref.get_array_of_samples())

filename = "new_2/03_27_01.wav"
file = AudioSegment.from_wav(filename)

print(len(file.get_array_of_samples()[:l]))

samples = file.get_array_of_samples()[:l]

result = AudioSegment(samples.tobytes(),
                    frame_rate=file.frame_rate,
                    sample_width=file.sample_width,
                    channels=file.channels)

result.export(filename, format = "wav")
```

```python
# MONO TO STEREO for WebMUSHRA purposes

from pydub import AudioSegment

filename = "new/03.wav"

left_channel = AudioSegment.from_wav(filename)
right_channel = AudioSegment.from_wav(filename)

stereo_sound = AudioSegment.from_mono_audiosegments(left_channel, right_channel)

stereo_sound.export(filename, format = "wav")
```

```python
# GENERATE ANCHORS

from pydub import AudioSegment
import pydub.scipy_effects #for low_pass_filter 

filename = "00"

order = 10
 
wav_file = AudioSegment.from_file(file = filename + ".wav", format = "wav")

freq = 3500

filtered = wav_file.low_pass_filter_cheby1(cutoff_freq=freq, order=order)
                
samples = filtered.get_array_of_samples()

result = AudioSegment(samples.tobytes(),
                    frame_rate=filtered.frame_rate,
                    sample_width=filtered.sample_width,
                    channels=filtered.channels)

file_name = filename + "_" + "35" + ".wav"
result.export(file_name, format = "wav")

# #################################################

freq = 7000

filtered = wav_file.low_pass_filter_cheby1(cutoff_freq=freq, order=order)
                
samples = filtered.get_array_of_samples()

result = AudioSegment(samples.tobytes(),
                    frame_rate=filtered.frame_rate,
                    sample_width=filtered.sample_width,
                    channels=filtered.channels)

file_name = filename + "_" + "70" + ".wav"
result.export(file_name, format = "wav")
```
